```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.

x_axis=["Jan", "Feb", "Mar", "April", "May", "June", 
        "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

"""Set the y-axis to a list of floats as the total fare in US dollars 
accumulated for each month"""

y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 
        43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax=plt.subplots()

"""Line above can also be written like this:
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)"""

# fig references the figure & ax references the axes of the figure


# Plot the data from the x and y axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add label for the legend
plt.plot(x_axis,y_axis, label="Boston", marker="^", color="blue",
        linewidth=2)

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis,y_axis,color="green",label="Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare By Month")

# Add the legend (label="Boston")
plt.legend()
```

```python
# Skill Drill 5.1.7
plt.scatter(y_axis,x_axis,color="red", label="Chicago")

# Add axes titles
plt.xlabel("Fare ($)")
plt.ylabel("Months")

# Add title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()

# Invert y-axis months using .gca() (get current axes)
plt.gca().invert_yaxis()
```

```python
# Make the ones further up on the y-axis bigger bubbles
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
# Make the bubbles bigger overall
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis,y_axis_larger)
```

```python
# This is also a way to make the bubbles bigger
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# Use of a new parameter (s)
fig,ax=plt.subplots()
ax.scatter(y_axis,x_axis,s=y_axis, label="Boston", c="skyblue",alpha=0.80, edgecolors="black",
          linewidth=2)

# Legend
plt.title("PyBer Fare by Month")

# Plot axes titles
plt.xlabel("Fare ($)")
plt.ylabel("Months")

# Add legend
plt.legend()
```

```python
# Create pie chart
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors=["slateblue","magenta","lightblue","green","yellowgreen","greenyellow",
       "yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values = (0, 0, 0.15, 0, 0, 0, 0.25, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis,
        counterclock=False,
        startangle=90,
        shadow=True,
        autopct='%.1f%%')

plt.show()
```

```python
# Get the standard deviation of the values in the y-axis
stdev=statistics.stdev(y_axis)
stdev
```

```python
plt.barh(x_axis,y_axis)
plt.xticks(np.arange(0,51,step=5.0))
plt.gca().invert_yaxis()
```

```python
# Object oriented approach
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure
fig,ax=plt.subplots(figsize=(8,8))
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))

# Create a minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
